```python
#library imports

from qiskit import QuantumCircuit, transpile
from qiskit_aer.primitives import Estimator
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
qc = QuantumCircuit(5,5)
qc.h(range(5))
print(qc.num_qubits)
qc.measure_all()
qc.draw(output='mpl')
```

```python
#Running on the backend

backend = AerSimulator()
transpiled_circuit = transpile(qc,backend)
backend_result = backend.run(transpiled_circuit,shots=10000).result()
shots = backend_result.get_counts()
print("Shots: ", shots)

```

```python
plot_histogram(shots)
```

```python
#finding the most likely binary password

password = sorted(shots.items(),key = lambda x:x[1],reverse = True) 
print("Your binary password is :", password[0][0])
```
